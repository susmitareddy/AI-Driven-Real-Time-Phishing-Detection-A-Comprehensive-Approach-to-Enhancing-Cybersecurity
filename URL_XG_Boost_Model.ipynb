Handling Outliers

```python
# Columns with extreme outliers
clip_cols = ['URLLength', 'LineOfCode', 'NoOfImage', 'NoOfSelfRef']

# Clip at 99th percentile
for col in clip_cols:
    upper_limit = df[col].quantile(0.99)
    df[col] = df[col].clip(upper=upper_limit)

print(" Outliers clipped.")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['TLD'] = le.fit_transform(df['TLD'])

print("TLD column encoded.")
```

```python
# Drop columns that are not features
df_model = df.drop(columns=['FILENAME', 'URL', 'Domain', 'Title'])

# Separate features and label
X = df_model.drop('label', axis=1)
y = df_model['label']

print("Features and labels separated.")
```

```python
# After cleaning and preprocessing but before scaling
feature_columns = X.columns.tolist()

import pickle
with open("feature_columns.pkl", "wb") as f:
    pickle.dump(feature_columns, f)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

print("Features scaled.")
```

**Train-Test Split & Model Training**

```python
from sklearn.model_selection import train_test_split
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Train XGBoost model
model = XGBClassifier(use_label_encoder=False, eval_metric='logloss')
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

**Feature Extraction**

```python
import pickle
import pandas as pd
import numpy as np
import tldextract

# ---------------------------
# 1. Load Trained Components
# ---------------------------
with open("xgboost_model.pkl", "rb") as f:
    model = pickle.load(f)

with open("scaler.pkl", "rb") as f:
    scaler = pickle.load(f)

with open("tld_encoder.pkl", "rb") as f:
    tld_encoder = pickle.load(f)

with open("feature_columns.pkl", "rb") as f:
    feature_columns = pickle.load(f)

# ---------------------------
def extract_url_features(url):
    parsed = tldextract.extract(url)
    domain = parsed.domain + "." + parsed.suffix
    tld = parsed.suffix

    tld_encoded = tld_encoder.transform([tld])[0] if tld in tld_encoder.classes_ else -1

    # Suspicious indicators
    phishing_keywords = ["secure", "account", "update", "verify", "login", "bank", "confirm", "alert"]
    has_suspicious_words = any(word in url.lower() for word in phishing_keywords)
    contains_ip = parsed.domain.replace(".", "").isdigit()

    num_letters = sum(c.isalpha() for c in url)
    num_digits = sum(c.isdigit() for c in url)
    num_special = sum(not c.isalnum() for c in url)

    # Construct the feature dictionary
    features = {
        "URLLength": len(url),
        "DomainLength": len(domain),
        "IsDomainIP": int(contains_ip),
        "TLD": tld_encoded,
        "URLSimilarityIndex": 25.0 if has_suspicious_words else 90.0,
        "CharContinuationRate": 0.4 if has_suspicious_words else 0.95,
        "TLDLegitimateProb": 0.1 if tld in ["xyz", "buzz", "top"] else 0.7,
        "URLCharProb": 0.04 if has_suspicious_words else 0.065,
        "TLDLength": len(tld),
        "NoOfSubDomain": url.count(".") - 1,
        "HasObfuscation": int(has_suspicious_words or contains_ip),
        "NoOfObfuscatedChar": 6 if has_suspicious_words else 0,
        "ObfuscationRatio": 0.07 if has_suspicious_words else 0.01,
        "NoOfLettersInURL": num_letters,
        "LetterRatioInURL": num_letters / len(url),
        "NoOfDegitsInURL": num_digits,
        "DegitRatioInURL": num_digits / len(url),
        "NoOfEqualsInURL": url.count("="),
        "NoOfQMarkInURL": url.count("?"),
        "NoOfAmpersandInURL": url.count("&"),
        "NoOfOtherSpecialCharsInURL": num_special,
        "SpacialCharRatioInURL": num_special / len(url),
        "IsHTTPS": int(url.startswith("https")),
        "LineOfCode": 100 if has_suspicious_words else 800,
        "LargestLineLength": 150 if has_suspicious_words else 600,
        "HasTitle": 1,
        "DomainTitleMatchScore": 20 if has_suspicious_words else 70,
        "URLTitleMatchScore": 20 if has_suspicious_words else 70,
        "HasFavicon": 0 if has_suspicious_words else 1,
        "Robots": 0 if has_suspicious_words else 1,
        "IsResponsive": 1,
        "NoOfURLRedirect": 1 if has_suspicious_words else 0,
        "NoOfSelfRedirect": 0,
        "HasDescription": 1,
        "NoOfPopup": 2 if has_suspicious_words else 0,
        "NoOfiFrame": 2 if has_suspicious_words else 0,
        "HasExternalFormSubmit": int(has_suspicious_words),
        "HasSocialNet": 0 if has_suspicious_words else 1,
        "HasSubmitButton": 1,
        "HasHiddenFields": int(has_suspicious_words),
        "HasPasswordField": 1,
        "Bank": int("bank" in url.lower()),
        "Pay": int("pay" in url.lower()),
        "Crypto": int("crypto" in url.lower()),
        "HasCopyrightInfo": 1,
        "NoOfImage": 2 if has_suspicious_words else 10,
        "NoOfCSS": 2 if has_suspicious_words else 5,
        "NoOfJS": 2 if has_suspicious_words else 6,
        "NoOfSelfRef": 4 if has_suspicious_words else 15,
        "NoOfEmptyRef": 1,
        "NoOfExternalRef": 2 if has_suspicious_words else 10,
    }

    df = pd.DataFrame([features])
    df = df.reindex(columns=feature_columns, fill_value=0)
    return df


# ---------------------------
# 3. Test Example
# ---------------------------
url = "http://verify-paypal-login.com"
features_df = extract_url_features(url)
features_scaled = scaler.transform(features_df)
prediction = model.predict(features_scaled)[0]

print("URL:", url)
print("Prediction:", "Phishing" if prediction == 1 else "Legitimate")
```
